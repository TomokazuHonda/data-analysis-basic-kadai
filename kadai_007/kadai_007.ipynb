```python
import pandas as pd
#csvファイルをデータフレームとして読み込む
df = pd.read_csv("sample_pandas_6.csv")

def tax(x):
  return x * 1.10

tax_series = df['単価'].apply(tax)
tax_series.name = "単価(税込み)"
pd.concat([df,tax_series],axis=1)

category_df = pd.read_csv('category.csv')

df = pd.merge(df,category_df[['商品番号','カテゴリー']],how='inner',on='商品番号')
df
```

```python
#カテゴリーの出現回数
df['カテゴリー'].value_counts()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib


x = df['カテゴリー'].value_counts().index
y = df['カテゴリー'].value_counts()

# 以下はx:200個、y:6個でエラー(xとyのデータの長さが一致しない)
# x = df['カテゴリー']
# y = df['カテゴリー'].value_counts()

#①カテゴリー列の要素の出現頻度をカウントして、その結果を棒グラフで表示する
plt.title("カテゴリー別出現頻度")
plt.bar(x, y)
```
